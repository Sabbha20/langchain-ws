### FAISS DB

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
doc_load = loader.load()
doc_load
```

```python
text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=30)
text_splitter
```

```python
docs = text_splitter.split_documents(doc_load)
docs
```

```python
embeddings = OllamaEmbeddings(model='mxbai-embed-large')
embeddings
```

```python
db = FAISS.from_documents(docs, embeddings)
db
```

```python
query1="What is the content is about?"
result_docs = db.similarity_search(query1)
result_docs
```

```python
result_docs[0].page_content
```

```python
query1="What is the outcome of the war?"
result_docs = db.similarity_search(query1)
result_docs
```
