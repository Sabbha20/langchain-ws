#### Finding similarity


```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df['embeddings'] = df['Words'].apply(lambda x: embeddings.embed_query(x))
df.to_csv("words_embeddings.csv")
```

```python
new_df = pd.read_csv("words_embeddings.csv")
new_df
```

```python
word = "Mango"
word_embedding = embeddings.embed_query(word)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
df[f"{word}_similarity"] = df['embeddings'].apply(lambda x: cosine_similarity([x], [word_embedding])[0][0])
df
```
